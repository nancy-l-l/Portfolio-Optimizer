**Introduction**

The covariance matrix represents the covariances between multiple variables, offering a comprehensive view of their interactions. It plays a critical role in portfolio diversification, risk analysis, and constructing efficient portfolios.

The covariance matrix is essential to portfolio management in the finance industry. By taking into account the covariances between various assets, it aids investors in building diversified portfolios. Investing in assets with low or negative covariances can lower portfolio risk overall and increase prospective rewards.



The covariance formula for a population is given by:

$ \text{Cov}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n} $


*   Cov(X,Y) represents the covariance between variables X and Y.
*   Xi and Yi are the individual data points of X and Y, respectively.
*   $\bar{X}$ and $\bar{Y}$ represent the means(averages) of X and Y, respectively.
*   Σ indicates the sum of the terms.
*   n represents the number of data points.

The degree to which two variables move in tandem is measured by covariance. We can measure the magnitude and direction of the linear correlations between variables by computing their covariances. Finding patterns in the data and comprehending how changes in one variable may effect another require the knowledge of this information.

````python
import yfinance as yf
import matplotlib.pyplot as plt

# Stock symbols
symbols = ['MSFT', 'AMZN', 'MCD', 'META']

# Fetching historical stock prices for each symbol
data = {}
for symbol in symbols:
    stock_data = yf.download(symbol, start='2022-01-01', end='2023-01-01')
    data[symbol] = stock_data['Close']

# Plotting the stock prices
plt.figure(figsize=(10, 6))
for symbol, prices in data.items():
    plt.plot(prices, label=symbol)

plt.title('Historical Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
````

````python
from scipy.optimize import minimize
import yfinance as yf
import pandas as pd
import numpy as np
import random


def calculate_covariance_matrix(stock_data):
    stock_df = pd.DataFrame(stock_data)
    returns = stock_df.pct_change().dropna()
    covariance_matrix = returns.cov()
    return covariance_matrix

symbols = ['MSFT', 'AMZN', 'MCD', 'META']
data = {}
for symbol in symbols:
  stock_data = yf.download(symbol, start='2022-01-01', end='2023-01-01')
  data[symbol] = stock_data['Close']

covariance_matrix = calculate_covariance_matrix(data)
print("Covariance Matrix:")
print(covariance_matrix)

def calculate_portfolio_balance(weights, covariance_matrix):
    portfolio_return = np.sum(random_weights * expected_returns)
    portfolio_volatility = np.sqrt(np.dot(random_weights.T, np.dot(covariance_matrix, random_weights)))
    return portfolio_return, portfolio_volatility

# Generate random weights
random_weights = np.random.rand(len(symbols))
random_weights /= np.sum(random_weights)  # Normalize to ensure weights sum to 1
expected_returns = np.array([np.mean(data[symbol]) for symbol in symbols])


# Calculate portfolio balance with random weights
random_portfolio_return, random_portfolio_volatility = calculate_portfolio_balance(random_weights, covariance_matrix)

# Print results for the random portfolio
print("Random Portfolio Return:", random_portfolio_return)
print("Random Portfolio Volatility:", random_portfolio_volatility)


# Define a function to minimize (negative Sharpe ratio)
def objective_function(weights, covariance_matrix):
    portfolio_return, portfolio_volatility = calculate_portfolio_balance(weights, covariance_matrix)
    return -portfolio_return / portfolio_volatility

# Set optimization constraints
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# Set bounds for individual stock weights (0 to 1)
bounds = tuple((0, 1) for _ in range(len(symbols)))

# Initial guess for the optimizer
initial_weights = np.ones(len(symbols)) / len(symbols)

# Run the optimizer
result = minimize(objective_function, initial_weights, args=(covariance_matrix,), method='SLSQP', bounds=bounds, constraints=constraints)

# Extract optimized weights
optimized_weights = result.x

# Calculate portfolio balance with optimized weights
optimized_portfolio_return, optimized_portfolio_volatility = calculate_portfolio_balance(optimized_weights, covariance_matrix)

# Print results for the optimized portfolio
print("Optimized Portfolio Return:", optimized_portfolio_return)
print("Optimized Portfolio Volatility:", optimized_portfolio_volatility)
print("Optimized Portfolio Weights:", optimized_weights)
````
